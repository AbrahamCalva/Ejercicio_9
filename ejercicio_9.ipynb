```python
plt.figure(figsize = (12, 6)) 

marcas_clase = [1, 2, 3, 4, 5, 6, 7, 8]
frecuencia = [7912, 3736, 1370, 1091, 1016, 445, 411, 405] 

plt.barh(marcas_clase, frecuencia, 
        height = 0.8, edgecolor = "k", 
        color = ["#881800", "#DAA310", "#AFDA10", "#10DA19", "#1026DA", "#DA1057", "#08D9F1", "#92B1A5"])

marcas_textos = ["Tsuru", "Pick up", "Sentra", "Jetta 4a. generación", "Bora", "Estaquillas 35 Ton", "Pointer", "Sedan"]

plt.gca().invert_yaxis()
plt.yticks(marcas_clase,marcas_textos, fontsize = 15) 
plt.ylabel("Tipos de coche", fontsize = 25) 
plt.xlabel("Número de robos", fontsize = 25) 
plt.title("Número de robos de autos en los primeros 7 meses del año 2010", fontsize = 40) 
plt.grid() 
plt.show() 
```

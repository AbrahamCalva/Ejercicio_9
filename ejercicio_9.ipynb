## Diagrama de barras 

```python
plt.figure(figsize = (12, 6)) 

marcas_clase = [1, 2, 3, 4, 5, 6, 7, 8]
frecuencia = [7912, 3736, 1370, 1091, 1016, 445, 411, 405] 

plt.barh(marcas_clase, frecuencia, 
        height = 0.8, edgecolor = "k", 
        color = ["#881800", "#DAA310", "#AFDA10", "#10DA19", "#1026DA", "#DA1057"])

marcas_textos = ["Tsuru", "Pick up", "Sentra", "Jetta 4a. generación", "Bora", "Estaquillas 35 Ton", "Pointer", "Sedan"]

plt.gca().invert_yaxis()
plt.yticks(marcas_clase,marcas_textos, fontsize = 15) 
plt.ylabel("Frecuancia absoluta", fontsize = 25) 
plt.xlabel("Marcas de clase", fontsize = 25) 
plt.title("Diagrama de barras", fontsize = 40) 
plt.grid() 
plt.show() 
```

## Diagrama de pastel

```python
datos = [7912, 3736, 1370, 1091, 1016, 445, 411, 405]
separaciones = [0, 0, 0, 0, 0, 0, 0, 0]

plt.pie(datos, explode = separaciones,
        colors = ["#881800", "#DAA310", "#AFDA10", "#10DA19", "#1026DA"],
        labels = ["Tsuru", "Pick up", "Sentra", "Jetta 4a. generación", "Bora", "Estaquillas 35 Ton", "Pointer", "Sedan"],
        autopct = "%0.1f%%",
        pctdistance = 0.6,
        counterclock = False,
        startangle = 180)

plt.show()
```
